# An algorithmic strategy for backtesting and trading with the indicator Parabolic Stop And Reverse (PSAR)

## Background to the PSAR

### Definition

To start understanding why this technical indicator may be useful, lets start by defining the mathematical relationship describing this. The PSAR parameter is pre-calculated for the time-step n+1, by the data at the time step n, and is defined as

$S_{n+1} = S_n + \alpha (E - S_n)$, 

where S is the PSAR value, E is the previous extreme point value and $\alpha$ is an acceleration factor, increasing for each direction jump it hits in a trend. 

### Theory and hypothesis

#### What is it and what does it define?

The PSAR indicator, is a indicator that is supposed to predict reversals of trends. A new extreme point (EP) is definied, once it is reached, contributing to further calculations. The acceleration factor increases over time, as further steps are taken from a given extreme point. 

#### Why may it be useful to generate trading signals?

It is a dynamic tool, adjusting its setpoint after each trend reversal, meaning that it can help a trader indicate where the trend reverses occur. It is particularly useful in strong trends, due to the difference between the extreme point and current SAR value being larger. The acceleration factor is continuously increasing, indicating that we are constantly betting for a trend reverse, making the reverse a sharper cut through the SAR value, meaning that the signals are more easily generated as the trend continuous, which is good to yield a lower risk.

#### Limitations

It is often good to use it with a strenght indicator, such as a RSI indicator. This indicator only defines trend, not the strength of it.
